```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data = pd.read_csv('Churn_Modelling.csv')
data.head()
```

```python
## Preprocessing

# Drop columns

data = data.drop(['RowNumber','CustomerId','Surname'], axis = 1)
data.head()
```

```python
# Encode categorical variables

label_encoder_gender = LabelEncoder()
data['Gender'] = label_encoder_gender.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
onehot_geo = OneHotEncoder(sparse_output=False)
onehot_geo.fit(data[['Geography']])
names = onehot_geo.get_feature_names_out(['Geography'])
geo_encoder = onehot_geo.transform(data[['Geography']])
```

```python
encoded_df = pd.DataFrame(geo_encoder,columns =names )
encoded_df
```

```python
# Combine columns

data = pd.concat([data.drop('Geography',axis = 1),encoded_df],axis= 1)
```

```python
data
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(label_encoder_gender,file)

with open('oneHot_encoder_geo.pkl','wb') as file:
    pickle.dump(onehot_geo,file)
```

```python
# Divide the data into dependent and independent

X = data.drop('Exited',axis = 1)
Y = data['Exited']
```

```python
# Train and Test Split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)

#Scaling 

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# Building ANN

model = Sequential([
    Dense(64, activation = 'relu',input_shape = (X_train.shape[1],)),
    Dense(32, activation = 'relu'),
    Dense(1, activation = 'sigmoid')]


)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss = "binary_crossentropy",metrics = ['accuracy'])
```

```python
early_stopping_callback = EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history = model.fit(
    X_train,Y_train,validation_data = (X_test,Y_test),epochs=100,
    callbacks=early_stopping_callback
)
```

```python
model.save('model.h5')
```
